```python
import numpy as np
a=np.random.random((10,1))
print(np.mean(a))
```

```python
a=np.random.rand(20,1)
b=np.std(a)
print("std deviation is ",b)
b=b**2
print('variance is',b)
```

```python
a=np.random.rand(10,20)
b=np.random.rand(20,25)
g=np.matmul(a,b)
print(g)
print(np.sum(g))
```

```python
a=np.random.rand(10,1)
print(a)
def func(x):
    return 1/(1+np.exp(-x))
b=func(a)
print(b)
```
